```python
# 4.Q Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories
# ANS>>
import numpy as np
import matplotlib.pyplot as plt

# Generate categories and corresponding numerical values
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [np.random.normal(loc=mu, scale=10, size=100) for mu in range(10, 50, 10)]  # Generate random numerical values for each category

# Plot the distribution using a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(values, labels=categories)
plt.xlabel('Category')
plt.ylabel('Numerical Value')
plt.title('Distribution of Numerical Values across Categories (Box Plot)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot the distribution using a violin plot (alternative visualization)
plt.figure(figsize=(8, 6))
plt.violinplot(values, showmeans=True)
plt.xticks(range(1, len(categories) + 1), categories)
plt.xlabel('Category')
plt.ylabel('Numerical Value')
plt.title('Distribution of Numerical Values across Categories (Violin Plot)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# 5.Q Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
# ANS>>
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic data with correlated features
np.random.seed(0)  # for reproducibility
data = np.random.multivariate_normal(mean=[0, 0], cov=[[1, 0.5], [0.5, 1]], size=1000)

# Create a DataFrame for the data
import pandas as pd
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2'])

# Compute the correlation matrix
correlation_matrix = df.corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 3.Q>> Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.
# ANS>>
'''np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)'''

# ANS>>

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sales Data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Pivot the data to create a sales matrix
sales_matrix = df.pivot_table(values='Sales', index='Month', columns='Day')

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(sales_matrix, cmap='YlGnBu', annot=True, fmt='.0f', cbar_kws={'label': 'Sales'})
plt.title('Variation in Sales across Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')
plt.show()
```

```python
# 4.Q  Using the given x and y data, generate a 3D surface plot to visualize the function, Z = sin(ROOT(X**2+Y**2))
''' x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)'''

# ans>>
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a 3D surface plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(x, y, z, cmap='viridis')

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot of Z = sin(sqrt(X^2 + Y^2))')

# Add color bar
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
# 1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
# ans>>
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import math
# Define the x values
x = list(range(0, 11))

# Compute the corresponding y values (sine of x)
y = [i for i in map(lambda x: round(math.sin(x), 2), x)]

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

```python
# 2.Q Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
# ans>>
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Randomly generate data
np.random.seed(42)
data = {
    'x': np.random.randint(1, 100, 50),
    'y': np.random.randint(1, 100, 50),
    'sizes': np.random.randint(5, 20, 50),  # Random sizes
    'colors': np.random.choice(['red', 'blue', 'green'], 50)  # Random colors
}

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add circle glyphs for scatter plot
p.circle('x', 'y', size='sizes', color='colors', alpha=0.6, source=data)

# Show the plot
show(p)

```

```python
# 3.Q Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
# fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
# counts = [20, 25, 30, 35]

# ans>>
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars with the counts of different fruits
p.vbar(x=fruits, top=counts, width=0.9)

# Set the y-axis label
p.yaxis.axis_label = "Counts"

# Show the plot
show(p)
```

```python
# 4.Q Create a Bokeh histogram to visualize the distribution of the given data.
# data_hist = np.random.randn(1000)
# hist, edges = np.histogram(data_hist, bins=30)

# ans>>
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Output to notebook
output_notebook()

# Create a new plot with a title and axis labels
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a histogram renderer
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="navy", line_color="white", alpha=0.5)

# Show the plot
show(p)
```
